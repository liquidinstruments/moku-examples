```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

## 1. Prepare the environment

```python
def extract_data_from_moku_1channel(filename):

    # Skip the metadata lines (first 12) and read the data
    df = pd.read_csv(filename, skiprows=12, comment='%', header=None)

    # Rename columns for clarity (optional)
    df.columns = ['Time (s)', 'Probe A (V)']

    # Extract Probe A values as a list
    #probe_a_values = df['Probe A (V)'].tolist()
    probe_a_values = df['Probe A (V)'].tolist()

    return probe_a_values
```

```python
def downsample(signal, factor):
    """
    Downsamples a signal by keeping the peak (max absolute) in each window.
    
    Parameters:
    - signal (list of float): Input signal with spikes.
    - factor (int): Downsampling factor (size of each chunk).

    Returns:
    - list of float: Downsampled signal with key spikes preserved.
    """
    if factor <= 0:
        raise ValueError("Downsampling factor must be a positive integer.")
    
    downsampled = []
    for i in range(0, len(signal), factor):
        window = signal[i:i + factor]
        if window:
            # Preserve the point with the largest absolute value (most spike-like)
            spike = max(window, key=abs)
            downsampled.append(spike)
    
    return downsampled
```

```python
def save_downsampled_chunks_to_csv(signal, window_size, output_filename):
    """
    Splits a signal into windowed chunks and saves them to a CSV file.
    
    Parameters:
    - signal (list of float): The input downsampled signal with anomalies.
    - window_size (int): The number of samples per window (frame).
    - output_filename (str): The path of the CSV file to write.
    
    Each row in the CSV will have:
    - 'sample_n': the index of the window
    - 'data': a comma-separated string of the window's signal values
    """
    if window_size <= 0:
        raise ValueError("window_size must be a positive integer.")
    
    num_windows = len(signal) // window_size
    print(f'Generating and saving {num_windows} frames from the time series with anomalies')

    with open(output_filename, "w", newline="") as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(["sample_n", "data"])  # Header

        for i in range(num_windows):
            start_idx = i * window_size
            end_idx = start_idx + window_size
            chunk = signal[start_idx:end_idx]

            writer.writerow([i + 1, ",".join(map(str, chunk))])
```

## 2. Load the file with data saved from the normal noisy sine signal, downsample it and save it for training the autoencoder

```python
downsampling_factor = 10
```

```python

# The following file has been saved using the Moku app. It saved normal frames, at a rate of 1MS/s
filepath_normal = "data_from_moku/normal_10_20250528.csv"
filename_output_normal = "training_folder/training_precision_normal_20250528_10.csv"
probe_a_values_normal = extract_data_from_moku_1channel(filepath_normal)
```

```python

downsampled_signal_tot = downsample(probe_a_values_normal, factor=downsampling_factor)
```

```python
# Let's plot the first 12000 datatpoints to confirm the signal is as expected
n_datapoints = 1000
plt.figure(figsize=(12, 4)) 
plt.title(f"Original normal frame - first {n_datapoints} points")
x_sample = probe_a_values_normal[:n_datapoints]
plt.plot(x_sample)
```

```python
plt.figure(figsize=(12, 4)) 
plt.title(f"Downsampled normal frame")
downsampled_signal_sample = downsample(x_sample, factor=downsampling_factor)
plt.plot(downsampled_signal_sample)
```

```python

plt.figure(figsize=(10, 4))
plt.plot(x_sample, label='Original', linewidth=1, alpha=0.4)
plt.title(f"Comparison Normal vs Downsampled on a normal frame")
plt.plot(np.linspace(0, len(x_sample), len(downsampled_signal_sample)), downsampled_signal_sample, 'o-', label='Downsampled', color='orange')
plt.legend(loc='lower right')
plt.grid(True)
plt.tight_layout()
plt.show()

print(f'The signal has been downsampled from {len(x_sample)} to {len(downsampled_signal_sample)} datapoints')
```


The downsampled signal is split into 100 datapoints frames and saved. The frames are not overalapped but this could be added to increase complexity in the training dataset.

```python
window_size = 100

save_downsampled_chunks_to_csv(downsampled_signal_tot, 100, filename_output_normal)
```

As we have saved the data in precision mode we will expect to see the anomaly averaged down.

```python
downsampled_signal_anomalies_sample = downsample(x_anomalies_sample,factor=downsampling_factor) 
downsampled_signal_anomalies_tot = downsample(probe_a_values_anomalies,factor=downsampling_factor)

print(f'Downsampled the signal with anomalies using a factor of {downsampling_factor} - The signal has been downsampled from {len(probe_a_values_anomalies)} to {len(downsampled_signal_anomalies_tot)} datapoints')
```

```python
# Plot the sample frame
plt.figure(figsize=(12, 4)) 
plt.title("Downsampled - frame with anomaly")
plt.plot(downsampled_signal_anomalies_sample)
```

```python
# Plot the comparison original - downsampled sample signals

plt.figure(figsize=(10, 4))
plt.title("Downsampled vs Original - sample frame with anomaly")
plt.plot(x_anomalies_sample, label='Original', linewidth=1, alpha=0.4)
plt.plot(np.linspace(0, len(x_anomalies_sample), len(downsampled_signal_anomalies_sample)), downsampled_signal_anomalies_sample, 'o-', label='Downsampled', color='orange')
plt.legend(loc='lower right')
plt.grid(True)
plt.tight_layout()
plt.show()

print(f'The sample signal has been downsampled from {len(x_anomalies_sample)} to {len(downsampled_signal_anomalies_sample)} datapoints')
```
